````python
from huggingface_hub import InferenceClient

client = InferenceClient(
    "mistralai/Mistral-7B-Instruct-v0.3",
    token="TOKEN_HERE",
)
````

````python
with open("prompts.csv", mode='r', newline='') as file:
    reader = csv.reader(file)
    headers = next(reader)  # Read the header
    data = [tuple(row) for row in reader]  # Read the data and convert each row to a tuple
````

````python
def strip_triple_backticks(s):
    if s.startswith('```'):
        s = s[3:]
    if s.endswith('```'):
        s = s[:-3]
    return s

predictions = []
for idx, row in enumerate(data):
    message = client.chat_completion(
    	messages=[{"role": "user", "content": row[1]}],
    	max_tokens=2048,
        temperature=0.2,
    	stream=False,
    )

    res = message.choices[0].message.content
    res = strip_triple_backticks(res)
    predictions.append(res)
    print(f"Done {idx+1}/{len(data)}")
````

````python
processed_preds = {}
for idx, pred in enumerate(predictions):
    pred = pred.strip()
    pred += f"\t----- bird -----\t{data[idx][0]}"
    processed_preds[str(idx)] = pred
````

````python
import json
with open("predict_mini_dev_mistral_SQLite.json", 'w') as json_file:
    json.dump(processed_preds, json_file, indent=4)
````
